```python
# dependent and independent variables
X = df.drop('Vu (KN)', axis = 1)
y = df['Vu (KN)']
```

```python
# train_test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 64)
print( X_train.shape)
print( X_test.shape)
print( y_train.shape)
print( y_test.shape)
```

```python
# check for outliers 
sns.set_style("white")
plt.figure(figsize=(20,30),facecolor='white')
plotnumber = 1

for column in X.columns:
    ax = plt.subplot(3,3,plotnumber)
    sns.boxplot((X[column]))
    plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
X_train
```

```python
X_normal = (X - X.min()) / (X.max() - X.min())
```

```python
X_normal
```

## Linear Regression model

```python
from sklearn.linear_model import LinearRegression
LReg = LinearRegression()
```

```python
LReg.fit(X_train ,y_train)
```

```python
LReg.get_params()
```

```python
y_pred_train = LReg.predict(X_train)
y_pred_test = LReg.predict(X_test)
```

```python
from math import sqrt
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

```python
print('The RMSE  on the testing dataset is: ',sqrt(mean_squared_error(y_test,LReg.predict(X_test))))
print('The RMSE  on the training dataset is: ',sqrt(mean_squared_error(y_train,LReg.predict(X_train))))
print('The MSE  on the testing dataset is: ',(mean_squared_error(y_test,LReg.predict(X_test))))
print('The MSE  on the training dataset is: ',(mean_squared_error(y_train,LReg.predict(X_train))))
print('The MAE  on the testing dataset is: ',mean_absolute_error(y_test,LReg.predict(X_test)))
print('The MAE  on the training daset is: ',mean_absolute_error(y_train,LReg.predict(X_train)))
print('The r2_score on the testing dataset is: ',r2_score(y_test,LReg.predict(X_test)))
print('The r2_score on the training dataset is: ',r2_score(y_train,LReg.predict(X_train)))
print('The MAPE  on the testing dataset is: ',mean_absolute_percentage_error(y_test,LReg.predict(X_test)))
print('The MAPE  on the training daset is: ',mean_absolute_percentage_error(y_train,LReg.predict(X_train)))
```

## Actual vs Predicted Capacity

```python
sns.set_style("white")
plt.figure(figsize=(10,10))
plt.title('Linear Regression')
plt.xlabel("Predicted Capacity (KN)")
plt.ylabel("Actual Capacity (KN)")
plt.scatter(y_test,y_pred_test)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.show()
```
